```python
# 라이브러리 불러오기
import os
import pandas as pd
import numpy as np

# 데이터 읽기
path_dir = "./data"
file_list = os.listdir(path_dir)
total_df = pd.DataFrame(columns=['title', 'summary', 'who', 'criteria', 'what', 'how', 'calls', 'sites','category'])

# category를 db에 맞게 int로 바꿔주기
def category_to_int(x, file):

    if x=="임신·출산": return 15
    elif x=="영유아" : return 0
    elif x=="아동·청소년" : return 1
    elif x=="청년" : return 2
    elif x=="중장년" : return 3
    elif x=="노년" : return 4
    elif x=='장애인' : return 5
    elif x=='한부모' : return 6
    elif x=='다문화(새터민)' : return 7
    elif x=='저소득층' : return 8
    elif x=="교육" : return 9
    elif x=="고용" : return 10
    elif x=="주거" : return 11
    elif x=="건강" : return 12
    elif x=="서민금융" : return 13
    elif x=="문화" : return 14
    else: return file.split('.')[0]
```

```python
# csv data 모두 합치기
for file in file_list:
    if file=='.DS_Store': continue
    csv_df = pd.read_csv(path_dir+"/"+file, engine='python',encoding='utf-8-sig') # utf-8-sig, euc-kr,cp949

    # category 변환
    csv_df['category'] = csv_df['category'].map(lambda x: category_to_int(x,file))
    total_df = pd.concat([total_df, csv_df])
```

```python
# NaN을 없애고 "빈칸"이란 이름을 부여
total_df = total_df.fillna("해당없음")
```

```python
# 중복을 없애고 각 복지에 대해서 welfare_id를 부여
unique_df = total_df.drop_duplicates('title', keep='first')
```

```python
unique_df = unique_df.rename_axis('welfare_id').reset_index()
```

```python
# total_df와 unique_df를 title을 기준으로 welfare_id 넣어주기
category_df = pd.merge(total_df[['title','category']], unique_df[['welfare_id','title']],how='left', on='title')
```
