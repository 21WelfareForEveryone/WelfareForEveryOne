```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd
import time

model_path = './output/training_sts'

embedder = SentenceTransformer(model_path)
# CPU 사용
#embedder = SentenceTransformer(model_path, device='cpu') 
```

```python
em_corpus = pd.read_csv("welfare_for_embedding.csv")
```

## title+summary embedding 

```python
corpus = em_corpus['full'].tolist()
```

```python
corpus
```

```python
import time
start = time.time()  # 시작 시간 저장
corpus_embeddings = embedder.encode(corpus, convert_to_tensor=True)
print("time :", time.time() - start)  # 현재시각 - 시작시간 = 실행 시간
```

```python
corpus_embeddings.size()
```

```python
import json
corpus_embeddings = corpus_embeddings.cpu() #GPU일 때 사용
em_se_numpy = corpus_embeddings.numpy()
for i in range(em_se_numpy.shape[0]):
    em_corpus.loc[i,'embedding'] = json.dumps(em_se_numpy[i,:].tolist())
```

```python
em_corpus.head()
```

```python
import copy
temp_full_corpus = copy.deepcopy(em_corpus)
```

```python
temp_full_corpus = temp_full_corpus.rename(columns={'embedding':'embedding_full'})
```

```python
temp_full_corpus.head()
```
